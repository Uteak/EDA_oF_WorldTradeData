## 셰계지도에 국가별 수출입 현황을 표시한 자료입니다.

마지막 함수를 실행시키고 국가명을 입력하시면 해당국가의 수출입 그래프가 표시됩니다.

```python
# import
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import folium
from folium import IFrame
import base64
import os
import json
import requests
from pandas.io.json import json_normalize
from numpy import pad
import matplotlib.image as mpimg
```

```python
# 나라 좌표 Dict 할당
location_dict = {
    'Australia' : [-24.86636, 134.48718],
    'China' : [34.612618948976916, 103.75025926830725],
    'Germany' : [51.11755198483869, 10.275667071811524],
    'India' : [23.07428624022239, 79.88110506108171],
    'Japan' : [36.66837630964112, 138.21634499324102],
    'Korea' : [37, 127],
    'Russia' : [62.45334143237126, 93.95547820929569],
    'Saudi Arabia' : [23.769067408710892, 45.2530890986929],
    'Singapore' : [1.2448578298792008, 103.81391014220462],
    'Taipei' : [24.11677747998721, 121.47257848639039],
    'USA' : [39.59123330879094, -101.49081886687664],
    'Viet Nam' : [14.310652944100811, 107.98895701299575]
}

# 나라 목록 할당
df_country = ['Australia', 'China', 'Germany', 'India', 'Japan', 'Korea', 'Russia', 'Saudi Arabia', 'Singapore', 'Taipei', 'USA', 'Viet Nam']

```

```python
def Problem_1_world_map(country_name):   
    filedir = './data2_plot1/' + country_name
    forder = os.listdir(filedir)
    # folium map 생성
    flmMap_export = folium.Map(location=[37.4810005593188, 33.373106073818576], tiles="OpenStreetMap", zoom_start=3)

    # Map에 Marker 형식으로 사진 삽입 반복문
    for files in forder:
        path = os.getcwd()
        file_name = files
        
        a = file_name.split('_')[-1]
        Buyer = file_name.split('_')[0]
        Seller = a.split('.')[0]
        ctry_name = Seller
        loc = location_dict.get(ctry_name)
        path2 = filedir + '/' + file_name
        pic = base64.b64encode(open(path2,'rb').read()).decode()

        image_tag = '<img src="data:image/jpeg;base64,{}">'.format(pic)
        iframe = folium.IFrame(image_tag, width=1010, height=580)
        popup = folium.Popup(iframe, max_width=996)

        mkr = folium.Marker(
                location=loc,
                tooltip='Total Import From {} ' .format(ctry_name),
                popup=popup,)

        mkr.add_to(flmMap_export)

    #html 형식으로 저장
    return flmMap_export


country_text = '국가번호를 입력해 주세요.(숫자만 입력)\n1. China\n2. Viet_Nam\n3. Singapore\n4. Taipei__Chinese\n5. Australia\n6. Russian_Federation\n7. Germany\n8. Japan\n9. United_States_of_America\n10. Saudi_Arabia\n11. India\n'
country_dic = {-1 : 'hi', 0 : 'Korea', 1 : 'China', 2 : 'Viet Nam', 3 : 'Singapore', 4 : 'Taipei', 5 : 'Australia', 6 : 'Russia', 7 : 'Germany', 8 : 'Japan', 9 : 'USA', 10 : 'Saudi Arabia', 11 : 'India' }
country_name = ''
country_num = -1

while country_dic[country_num] not in df_country:
    country_num = int(input(country_text))
    country_name = country_dic[country_num]
    
Problem_1_world_map(country_name)
    
```
